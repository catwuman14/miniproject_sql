```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

url = 'https://www.worldometers.info/world-population/myanmar-population/'
```

```python
response = requests.get(url)
response 
```

```python
soup = BeautifulSoup(response.content, 'html.parser')
print(soup.prettify())
```

```python
tbodies = soup.find_all("tbody")
tbodies
```

```python
city_element = soup.find_all("td", style = "font-weight: bold; font-size:17px; text-align:left; padding-left:10px; padding-top:5px; padding-bottom:5px")
city_element
```

```python
city = [i.get_text() for i in city_element]
city
```

```python
city_population = soup.find_all("td", style = " font-size:17px; text-align:left; padding-left:10px; padding-top:5px; padding-bottom:5px")
population = [i.getText() for i in city_population]
population
```

```python
df = pd.DataFrame ({"City":city, "Population":population})
df
```

```python
df.dtypes
```

```python
df['Population'] = df['Population'].map(lambda x: x.replace(',', ''))
```

```python
df['Population'] = df['Population'].apply(int)
```

```python
df.dtypes
```

```python
df.to_csv('/Users/catherinewu/Downloads/myanmarpop.csv')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "payment_type_city (1).csv"
df = pd.read_csv(file_path)

# Display column names for verification
print("Columns in CSV:", df.columns)

# Automatically handle column names if different
expected_columns = ["city", "payment_type", "count"]
if len(df.columns) != len(expected_columns):
    df.columns = df.columns[:len(expected_columns)]

df.columns = expected_columns

# Aggregate the count data by payment type
overall_data = df.groupby("payment_type")[["count"]].sum().reset_index()

# Generate a single pie chart for overall payment type distribution
plt.figure(figsize=(8, 8))
plt.pie(overall_data["count"], labels=overall_data["payment_type"], autopct='%1.1f%%', startangle=140)
plt.title("Overall Payment Type Distribution")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file
file_path = "payment_type_city (1).csv"
df = pd.read_csv(file_path)

# Display column names for verification
print("Columns in CSV:", df.columns)

# Automatically handle column names if different
expected_columns = ["city", "payment_type", "count"]
if len(df.columns) != len(expected_columns):
    df.columns = df.columns[:len(expected_columns)]

df.columns = expected_columns

# Get unique cities
cities = df["city"].unique()

# Create subplots
fig, axes = plt.subplots(1, len(cities), figsize=(15, 5))

# Ensure we handle cases where there's only one city
if len(cities) == 1:
    axes = [axes]

# Generate a pie chart for each city
for ax, city in zip(axes, cities):
    city_data = df[df["city"] == city]
    ax.pie(city_data["count"], labels=city_data["payment_type"], autopct='%1.1f%%', startangle=140)
    ax.set_title(f"Payment Type Distribution in {city}")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create the DataFrame
data = {
    'city': ['Mandalay', 'Mandalay', 'Mandalay', 'Nay Pyi Taw', 'Nay Pyi Taw', 'Nay Pyi Taw', 'Yangon', 'Yangon', 'Yangon'],
    'payment': ['Cash', 'Credit card', 'Ewallet', 'Cash', 'Credit card', 'Ewallet', 'Cash', 'Credit card', 'Ewallet'],
    'total_spend': [35337, 37349, 33513, 43085, 30326, 37158, 33780, 33097, 39325],
    'percentage_of_total': [33.2743, 35.1689, 31.5568, 38.9666, 27.4272, 33.6062, 31.8073, 31.1642, 37.0285]
}

df = pd.DataFrame(data)

# Pivot the data to get 'total_spend' as columns for each payment method
pivot_df = df.pivot(index='city', columns='payment', values='total_spend')

# Plotting the bar chart
pivot_df.plot(kind='bar', stacked=False, figsize=(10, 6), color=['#1F77B4', '#FF7F0E', '#2CA02C'])
plt.title('Total Spend by City and Payment Method')
plt.xlabel('City')
plt.ylabel('Total Spend')
plt.xticks(rotation=45)
plt.legend(title='Payment Method')
plt.tight_layout()

# Show the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the CSV file
file_path = "product_lines_gender.csv"
df = pd.read_csv(file_path)

# Rename columns for easier access
df.columns = ["product_line", "customer_type", "gender", "total_spend"]

# Split data into members and normal shoppers
df_members = df[df["customer_type"] == "Member"]
df_normal = df[df["customer_type"] == "Normal"]

# Pivot the data for better visualization
df_pivot_members = df_members.pivot(index="product_line", columns="gender", values="total_spend")
df_pivot_normal = df_normal.pivot(index="product_line", columns="gender", values="total_spend")

# Create subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 6), sharey=True)

# Plot Members data
df_pivot_members.plot(kind="bar", width=0.8, ax=axes[0])
axes[0].set_title("Spend by Gender for Members")
axes[0].set_xlabel("Product Line")
axes[0].set_ylabel("Total Spend")
axes[0].tick_params(axis='x', rotation=45)
axes[0].grid(axis='y', linestyle='--', alpha=0.7)

# Plot Normal shoppers data
df_pivot_normal.plot(kind="bar", width=0.8, ax=axes[1])
axes[1].set_title("Spend by Gender for Normal Shoppers")
axes[1].set_xlabel("Product Line")
axes[1].tick_params(axis='x', rotation=45)
axes[1].grid(axis='y', linestyle='--', alpha=0.7)

plt.tight_layout()
plt.show()
```
